The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ... $$

Let us list the factors of the first seven triangle numbers:

- 1: 1
- 3: 1,3
- 6: 1,2,3,6
- 10: 1,2,5,10
- 15: 1,3,5,15
- 21: 1,3,7,21
- 28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

[Problem](https://projecteuler.net/problem=12)

```python
import time
import math

def factorer(num):
    counter=0
    for i in range(1, math.ceil(math.sqrt(num))):
        if num % i == 0:
            if isinstance(math.sqrt(num), int):
                counter+=1
            else:
                counter+=2
    return counter
```

```python
start = time.time()

tri_num = 0
counter = 0
factors = 0
while factors < 500:
    counter += 1
    tri_num = tri_num + counter
    factors = factorer(tri_num)
    elapsed = time.time()-start
print('Counter: '+ str(counter),'\n','Triangular Number: '+str(tri_num),'\n', 'Time Taken: '+str(elapsed),'\n','\n')
```

```python
a=1000000000000
start = time.time()
print(factorer(a),time.time()-start)
```
