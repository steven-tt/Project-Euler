Pentagonal numbers are generated by the formula, $P_n=\frac{n(3n-1)}{2}$. The first ten pentagonal numbers are:

$$1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...$$

It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, $70 − 22 = 48$, is not pentagonal.

Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k − P_j|$ is minimised; what is the value of D?

[Problem](https://projecteuler.net/problem=44)

### Things from wiki
[Pentagonal numbers wiki](https://en.wikipedia.org/wiki/Pentagonal_number#Tests_for_pentagonal_numbers)

Given a positive integer $x$, to test whether it is a (non-generalized) pentagonal number we can compute

$${\displaystyle n={\frac {{\sqrt {24x+1}}+1}{6}}.} $$

The number $x$ is pentagonal if and only if $n$ is a natural number. In that case $x$ is the $n^{th}$ pentagonal number.

```python
import time #for timing
import math # for sqruare rooting
```

```python
def pentagonal(x):
    """
    Tests to see if the input x is a pentagonl number using the inverse function
    from the wikipedia article
    """
    n=(math.sqrt(24*x+1)+1)/(6)
    if n == int(n):
        return True

start = time.time() # starting time

notfound = True # lets the while loop go
i=0 # int variable for counter
while notfound:
    i+=1 # counter
    n=(i*(3*i-1))/2 # pentagonal number from index i
    for j in range(i,0,-1): # we know that i !=j and we can't include 0
        m=(j*(3*j-1))/2 #pentagonal number from index j
        if pentagonal(n-m) and pentagonal(n+m): #tests if n and m are pentagonal additivly and subtactily (pretty sure made up words)
            print('Our two pentagonal numbers are n={} and m={}. The result of n-m={}.'.format(int(n),int(m),int(n-m)))
            print('The calculation took {} seconds. We had to check {} numbers (I think) to get our solution.'.format(round(time.time()-start,2),int(i*(i-j))))
            # this is broken into two prints just because it looks prettier to read on my small screen
            notfound = False #stops the while loop
            break #stops the for loop
```
