Consider all integer combinations of $a^b$ for $2 ≤ a ≤ 5$ and $2 ≤ b ≤ 5$:

$2^2=4$, $2^3=8$, $2^4=16$, $2^5=32$ <br>
$3^2=9$, $3^3=27$, $3^4=81$, $3^5=243$ <br>
$4^2=16$, $4^3=64$, $4^4=256$, $4^5=1024$<br>
$5^2=25$, $5^3=125$, $5^4=625$, $5^5=3125$<br>
If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125

How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?

```python
import time
num_of_cal=0
start = time.time()
set_of_numbers =set() # Makes an empty set
for a in range(2,101):
    for b in range(2,101):
        set_of_numbers.add(a**b)
        num_of_cal+=1
elapsed = time.time()-start
print(f'The number of distinct terms in the sequence is {len(set_of_numbers)}\nThis took {elapsed:1.3f} seconds\nThere were {num_of_cal} calculations done to get this solution')
```
